```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
browser=webdriver.Firefox()
```

```python
browser.get(url)
```

```python
html=browser.page_source
html
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
table2=soup.select("figure.wp-block-table.is-style-stripes")[1]
```

```python
table2.find_all('th')
```

```python
thead=[i.text.strip() for i in table2.find_all('th')]
thead
```

```python
table2.find_all('tbody')[0].text
```

```python
soup.select("figure.wp-block-table.is-style-stripes")[1].tbody.find_all('tr')[0]
```

```python
data=[]
for r in soup.select("figure.wp-block-table.is-style-stripes")[1].tbody.find_all('tr'):
    data.append([i.text.strip() for i in r.find_all('td')])
```

```python
data
```

```python
import pandas as pd
df=pd.DataFrame(data)
df.columns=thead
df
```

```python
def shark_table():
    thead = [i.text.strip() for i in table2.find_all('th')]
    data=[]
    for r in soup.select('figure.wp-block-table.is-style-stripes')[1].tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```

```python
shark_table()
```

```python
df.shape
```

```python
df.to_csv("shark_tank.csv")
```
